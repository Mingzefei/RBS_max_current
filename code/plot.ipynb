````python
def get_data(filename):
    # read data from e2f2-Greedy.txt
    with open('./raw-results/'+filename+'.txt', 'r') as f:
        lines = f.readlines()
        lines = [line.strip() for line in lines]
        lines = [line.split(' ') for line in lines]
        lines = [[float(x) for x in line] for line in lines]
        data = np.array(lines)
        data = data.T.flatten()
    # change data to an increasing sequence
    for i in range(1, len(data)):
        data[i] = data[0:i+1].max()
    # add 0 at the beginning
    data = np.insert(data, 0, 0)
    return data
````

````python
data = {
    'Greedy': [],
    'SA': [],
    'GA': []
}
for file in ['e2f2-Greedy','e2f2-SA','e2f2-GA']:
    data[file.split('-')[1]] = get_data(file)
# plot
fig = plt.figure(figsize=(3.5,3))
# add line y=2. and text 'correct value'
plt.plot([-200,4000],[2,2], '--', color='r')
plt.text(500, 2.2, 'Correct value=2', color='r',size=12)
plt.plot(data['Greedy'], label='Greedy')
plt.plot(data['SA'], label='SA')
plt.plot(data['GA'], label='GA')
plt.xlabel('Iteration')
plt.ylabel('Objective value')
plt.legend()
plt.ylim(-0.2,6)
plt.xlim(-200,4000)

# save
plt.savefig('../attachments/e2f2-algorithm-comparison.png')
````
